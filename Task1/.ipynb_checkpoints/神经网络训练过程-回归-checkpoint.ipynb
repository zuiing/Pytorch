# 关系拟合(回归)

```python
#导入包
%matplotlib inline
import torch
import torch.nn.functional as F # 激励函数都在这儿 0_0
import matplotlib.pyplot as plt
from torch.autograd import Variable
```

##  建立数据集

创建一些假数据来模拟真实的情况 比如一个一元二次函数：y=a*x^2+b,并给数据加上一点噪声。

```python
x = torch.unsqueeze(torch.linspace(-1, 1, 100), dim=1)  # x data (tensor), shape=(100, 1)  unsqueeze将x从一维变成二维数据，因为Variable只能处理二维数据
y = x.pow(2) + 0.2*torch.rand(x.size())                 # noisy y data (tensor), shape=(100, 1)

x, y = Variable(x), Variable(y)

# 画出原始数据的散点图
plt.scatter(x.data.numpy(), y.data.numpy())
plt.show()
```

##  建立神经网络

建立一个神经网络可以直接实验torch中的体系。先定义所有的层属性(__init__()),然后再一层层搭建(forward())层与层的关系链接。建立关系时，会用到激励函数...

```python
class Net(torch.nn.Module): # 继承torch的Module
    def __init__(self,n_feature,n_hidden,n_output): # 类似于构造函数
        super(Net,self).__init__() # 继承__init__()函数功能
        #定义每层的函数形式
        self.hidden=torch.nn.Linear(n_feature,n_hidden) # 隐藏层线性输出
        self.predict=torch.nn.Linear(n_hidden,n_output) # 输出层线性输出
        
    def forward(self,x): # 重写Module中的forward功能
        #正向传播输入值，神经网络分析出输出值
        x=F.relu(self.hidden(x)) # 激励函数(隐藏层的线性值)
        x=self.predict(x) # 输出值
        return x
  
net=Net(n_feature=1,n_hidden=10,n_output=1)

print(net) # 打印net的结构
```

## 训练网络

```python
# optimizer是训练的工具，即优化器
optimizer=torch.optim.SGD(net.parameters(),lr=0.2) # 传入net的所有参数和学习率
loss_func=torch.nn.MSELoss() # 预测值和真实值的误差计算

for t in range(200):
    prediction=net(x) # 喂给net训练数据x，输出预测值
    loss=loss_func(prediction,y) # 计算预测值和真实值的误差
    
    optimizer.zero_grad() # 清空上一步的参与更新参数值
    loss.backward() # 误差反向传播，计算参数更新值
    optimizer.step  # 将参数更新值施加到net的parameters上

# 训练结果显示
plt.scatter(x.data.numpy(),y.data.numpy())
plt.plot(x.data.numpy(),prediction.data.numpy(),'r-',lw=5)
plt.text(0.5, 0, 'Loss=%.4f' % loss.data.numpy(), fontdict={'size': 20, 'color':  'red'})
plt.show()
```
